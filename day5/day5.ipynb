```python
import copy
```

```python
with open('real_input.txt') as day5_input:
    input_day_5_list = []
    for line in day5_input:
        line = line.strip()
        if line != '':
            input_day_5_list.append(line)
       
            
        
       
            
```

```python
def is_source_in_line(source_range_start, range_length, current_source):
    source_in_line = False
    source_range_end = source_range_start + range_length
    if current_source >= source_range_start and current_source < source_range_end:
        source_in_line = True 
    return source_in_line
    

    
```

```python
def update_destination_value_if_not_default(source_range_start, destination_range_start, current_source, mapping_dict):
    destination_index = current_source - source_range_start

    current_destination = destination_index + destination_range_start
   
    mapping_dict[copy.deepcopy(current_source)] = current_destination
    



```

```python
def update_destination_value_if_default(mapping_dict):
    for key, value in mapping_dict.items():
        if value == '':
            mapping_dict[key] = copy.deepcopy(key)
```

```python
def create_map(input_list):
    seeds_list = input_list[0].split()[1:]
    mappings_list = input_list[1:]
    mapping_dict = {key:int(value) for key,value in enumerate(seeds_list)}
    for line in mappings_list:
        if 'map' in line:
            
            mapping_dict = {key:'' for key in mapping_dict.values()}
        else: 
            
            numbers_list = line.split()
            destination_range_start = int(numbers_list[0])
            source_range_start = int(numbers_list[1])
            range_length = int(numbers_list[2])
            for key in mapping_dict.keys():
                
                if is_source_in_line(source_range_start, range_length, key):
                    
                    update_destination_value_if_not_default(source_range_start, destination_range_start, key, mapping_dict)
                    
                    
           
                
            update_destination_value_if_default(mapping_dict)

           
                
    
    return(mapping_dict)
        
    
              



            
    

        

    
    
```

```python
map_dict = create_map(input_day_5_list)
```

```python
def is_range_in_line(source_range_start, range_length, current_source_range):
    range_in_line = False
    source_range_end = source_range_start + range_length - 1
    if current_source_range[1] < source_range_start or current_source_range[0] > source_range_end:
        range_in_line = False
    else:
        range_in_line = True 
    return range_in_line
```

```python
def update_range_keys(mapping_dict, source_range_start, range_length, key):
    source_range_end = source_range_start + range_length
    mapping_dict_copy = copy.deepcopy(mapping_dict)
    if key[0] <= source_range_start:
        del mapping_dict_copy[key]

        mapping_dict_copy[(key[0], source_range_start - 1)] = ''
        mapping_dict_copy[(source_range_start, source_range_end)] = ''
        
        if key[1] > source_range_end:


            mapping_dict_copy[(key[0], source_range_start - 1)] = ''
            mapping_dict_copy[(source_range_start, source_range_end)] = ''
            mapping_dict_copy[(source_range_end + 1, key[1])] = ''
            
    if key[0] > source_range_start:
        if key[1] >= source_range_end:
            try: del mapping_dict_copy[key]
            except: print(f'dict causing key error: {mapping_dict_copy}, key: {key}')
            mapping_dict_copy[(key[0], source_range_end)] = ''
            mapping_dict_copy[(source_range_end + 1, key[1])] = ''
            
    return mapping_dict_copy


    
```

```python
def update_values(source_range_start, destination_range_start, range_length, mapping_dict, key):
    
    destination_index = key[0] - source_range_start
    new_start = destination_range_start + destination_index
    new_end = new_start + key[1] - key[0]
   
    mapping_dict[copy.deepcopy(key)] = (new_start, new_end)
```

```python
def create_map_part_2(input_list):
    seeds_list = input_list[0].split()[1:]
    seed_range_start_list = [int(item) for item in seeds_list if seeds_list.index(item)%2 == 0]
    seed_range_end_list = [int(item) for item in seeds_list if seeds_list.index(item)%2 == 1]
    seed_range_list = [(start, start + end) for start, end in zip(seed_range_start_list, seed_range_end_list)]
    mapping_dict = {key: value for key, value in enumerate(seed_range_list)}
   
    
    mappings_list = input_list[1:]

    
    for line in mappings_list:
        
        if 'map' in line:
            
            
            mapping_dict = {key:'' for key in mapping_dict.values()}
            
        else: 

            
            
            numbers_list = line.split()
            destination_range_start = int(numbers_list[0])
            source_range_start = int(numbers_list[1])
            range_length = int(numbers_list[2])
            keys_list = [key for key in mapping_dict.keys()]
            keys_list_copy = copy.deepcopy(keys_list)
            
            for key in keys_list_copy:
                
                
                if is_range_in_line(source_range_start, range_length, key):

                    
                   
                    mapping_dict = update_range_keys(mapping_dict, source_range_start, range_length, key)
                   
            
                    
                   

           
            
            for key in mapping_dict.keys():
                if is_range_in_line(source_range_start, range_length, key):

                    
                    
                    update_values(source_range_start, destination_range_start, range_length, mapping_dict, key)

                    
            
       
            
            update_destination_value_if_default(mapping_dict)
            
          

               
            

           
                
    
    return(mapping_dict)
```
