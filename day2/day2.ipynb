## Part 1

```python
with open('real_input.txt') as day2_input:
    input_day_2_list = []
    for line in day2_input:
        input_day_2_list.append(line.split())

    
```

```python
def calculate_number_combinations(input_list, no_red = 12, no_green = 13, no_blue = 14):
    total = 0
    for index1, game in enumerate(input_list): 
        too_many_red = False
        too_many_green = False 
        too_many_blue = False
        substring = game[2:]
        for index2, item in enumerate(substring):
            if 'red' in item and int(substring[index2-1]) > no_red:
                too_many_red = True
            if 'green' in item and int(substring[index2-1]) > no_green:
                too_many_green = True
            if 'blue' in item and int(substring[index2-1]) > no_blue:
                too_many_blue = True
        if too_many_red == False and too_many_green == False and too_many_blue == False:
            total += index1 + 1
    return total
```

```python
calculate_number_combinations(input_day_2_list)
```

```python
def calculate_minumum_combination(input_list):
    total = 0
    for game in input_list: 
        no_red = 0
        no_green = 0
        no_blue = 0
        substring = game[2:]
        for index2, item in enumerate(substring):
            if 'red' in item and int(substring[index2-1]) > no_red:
                no_red = int(substring[index2-1])
            if 'green' in item and int(substring[index2-1]) > no_green:
                no_green = int(substring[index2-1])
            if 'blue' in item and int(substring[index2-1]) > no_blue:
                no_blue = int(substring[index2-1])
        
        total += no_red * no_green * no_blue
    return total
```

```python
calculate_minumum_combination(input_day_2_list_part_2)
```
